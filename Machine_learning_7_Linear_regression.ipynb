```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("main.csv")
df.head()

GRE_Score = df['GRE Score'].tolist()
Chances_Of_Admit = df['Chance of Admit '].tolist()
fig = px.scatter(x = GRE_Score,y = Chances_Of_Admit,color = GRE_Score)
fig.show()
```

```python
m =  0.01
c =  -2.45
y = []

for x in GRE_Score:
  y_value = m * x + c
  y.append(y_value)

fig = px.scatter(x = GRE_Score,y = Chances_Of_Admit)
fig.update_layout(shapes = [
    dict(
        type = 'line',
        y0 = min(y),
        y1 = max(y),
        x0 = min(GRE_Score),
        x1 = max(GRE_Score)
    )
])
fig.show()
```

```python
x =  600
y = m * x + c
print(f"hances of getting admitted in the university is {y} with a score of {x}")
```

```python
import numpy as np

GRE_numpy_list = np.array(GRE_Score)
Chances_numpy_list = np.array(Chances_Of_Admit)

m, c  = np.polyfit(GRE_numpy_list,Chances_numpy_list,1)
y = []
for x in GRE_numpy_list:
  y_value = m * x + c
  y.append(y_value)

fig = px.scatter(x = GRE_numpy_list,y = Chances_numpy_list)
fig.update_layout(shapes=[
    dict(
        type = 'line',
        y0 = min(y),
        y1 = max(y),
        x0 = min(GRE_numpy_list),
        x1 = max(GRE_numpy_list)
    )
])
fig.show()
print(type(GRE_numpy_list))
```

```python
x = 600
y = m*x + c
print(f'for a score of {x} chances of admit in the uni is {y}')
```
